```python
#import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Funciones auxiliares

```python
def pritnTabMax(Data,namesIMGE):
    n = Data.shape[0]
    for i in range(n):
        print(namesIMGE[i], ' & ',Data[i,0] , ' & ' ,Data[i,1] , ' & ', Data[i,2] , ' \\\\' )
```

```python
def printImgs(images, n=1, m=1, title=''):
    
    images = images[:n*m]
    fig, axs = plt.subplots(
        nrows=n, ncols=m, 
        figsize=(18, 20), 
        subplot_kw={
            'xticks': [], 
            'yticks': []
        }
    )
    for ax, image in zip(axs.flat, images):
        ax.imshow(image, cmap='gray')
    plt.savefig(title)    
    plt.show()
```

```python
def printGrafs(images, n=1, m=1, title=''):
    
    images = images[:n*m]
    fig, axs = plt.subplots(
        nrows=n, ncols=m, 
        figsize=(18, 20), 
        subplot_kw={
            'xticks': [], 
            'yticks': []
        }
    )
    for ax, image in zip(axs.flat, images):
        ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.savefig(title)    
    plt.show()
```

# Programa principal

```python
LAMvaltxt = np.array(['0.000001','0.000003','0.000005','0.000007'])
```

```python
#Lectura

ruta  = './txt83CG/'
ruta2 = './restoration/P83CG/BestPnsr'

ls = []
lsgraf = []
resData = []

k = 4 #Lambda =0
namesIMGE = np.array(['F16_GT.png','image_Baboon512rgb.png', 'image_House256rgb.png','image_Lena512rgb.png','image_Peppers512rgb.png'])

for i in range(len(namesIMGE)):
    nameimg =  namesIMGE[i]
    

    name = nameimg.replace('.png','.csv')

    fname = ruta + 'P83' + 'denGaussMAXval' + name
    maxPnsr = np.genfromtxt(fname, delimiter=',')
    
    fname = ruta + 'P83' + 'denGaussPNSR' + name 
    plotY = np.genfromtxt(fname, delimiter=',')
    
    Y = np.array(plotY).T
    
    #Tabla
    tabMax = np.array([ maxPnsr[k,1],maxPnsr[k,2], Y[k][-1] ])
    resData.append(tabMax)
    print(tabMax)
    
    #Grafica
    x = np.arange(Y.shape[1])
    plt.plot(x,Y[k])
    plt.xlabel('Iteracin')
    plt.ylabel('PNSR')
    plt.title(nameimg)
    namefig = './outGraf/'+ str(i) +'lamb' + LAMvaltxt[k] + "grafPNSR83CGE.png"
    plt.savefig(namefig)
    plt.show()
    
    #Imagen grafica
    imgOut = cv2.imread(namefig)
    lsgraf.append(imgOut)
    
    #Imagen
    fname = ruta2 + LAMvaltxt[k] + nameimg
    imgOut = cv2.imread(fname, cv2.IMREAD_GRAYSCALE)
    ls.append(imgOut)
```

```python
tab = np.array(resData) 
```
